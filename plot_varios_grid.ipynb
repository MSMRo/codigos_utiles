```python
import glob
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import os
```

```python
path = "/kaggle/input/cap-sleep-unbalanced-dataset/Balanced Data/Balanced Data"

path_ls = os.listdir(path)
path_ls
```

## Lectura de los csv

```python
df1 = pd.read_csv(path + '/'+path_ls[0], header=None)
df1.head()
```

```python
df1.info()
```

```python
nfig = 15
# nrows x ncols = nfig
nrows = 5
ncols = 3

fig, ax = plt.subplots(nrows=nrows, ncols=ncols ,figsize=(10,10))
fig.suptitle(f"{path_ls[0]}", fontsize=18)
fig.tight_layout()

for n in range(nfig):
    row = n//ncols
    col = n%ncols
    
    ax[row,col].plot(df1.iloc[n,:-1])
    ax[row,col].grid(ls=":")
    ax[row,col].set_ylim(-50,50)
```

```python
def group_plot(nfig, nrows, ncols, df_array,title_plot="titulo xxx", ylim1=-10, ylim2=10):
    
    if nrows*ncols != nfig:
        raise "Error: nrows*ncols != nfig"


    fig, ax = plt.subplots(nrows=nrows, ncols=ncols ,figsize=(10,10))
    fig.suptitle(f"{title_plot}", fontsize=18)
    fig.tight_layout()

    for n in range(nfig):
        row = n//ncols
        col = n%ncols

        ax[row,col].plot(df_array.iloc[n,:])
        ax[row,col].grid(ls=":")
        ax[row,col].set_ylim(ylim1,ylim2)
```

```python
group_plot(15,5,3,df1.iloc[100:115,:-1], title_plot="100:115", ylim1=-50, ylim2=50)
```
